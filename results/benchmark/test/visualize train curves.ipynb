```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
func_names = ["id", "gaussian1",  "exp", "sin", "f1", "f2"]
train_losses = dict()
for func_name in func_names:
    fn = "train_curve_{}.npy".format(func_name)
    array = np.load(fn, allow_pickle=True)
    train_losses[func_name] = array
```

```python
for func_name in func_names:
    plt.plot(train_losses[func_name])
    plt.title(func_name);
    plt.show()
```
