```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
folder = "m_1_exp_1/"
func_names = ["gaussian1"]
train_losses = dict()
for func_name in func_names:
    fn = folder + "train_curve_{}.npy".format(func_name)
    array = np.load(fn, allow_pickle=True)
    train_losses[func_name] = array
for func_name in func_names:
    plt.plot(train_losses[func_name])
    plt.title(func_name);
    plt.show()
```
