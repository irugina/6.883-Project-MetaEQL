```python
plt.figure(figsize = (10,5))
plt.ylim(top=40)  # adjust the top leaving bottom unchanged
plt.ylim(bottom=0)  # adjust the bottom leaving top unchanged
folder = "m_1_exp_5/"
func_names = ["id", "gaussian1",  "exp", "sin", "f1"]
train_losses = dict()
for func_name in func_names:
    fn = folder + "train_curve_{}.npy".format(func_name)
    array = np.load(fn, allow_pickle=True)
    train_losses[func_name] = array
for func_name in func_names:
    plt.plot(train_losses[func_name][0:35], label = func_name)
plt.title("5 symbolic regression tasks", fontsize=17);
plt.legend(fontsize=15)
plt.tick_params(labelcolor='black', labelsize='large', width=2, length=10)
plt.savefig('m_1_5_tasks')
```
